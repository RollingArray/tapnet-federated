```python
import os

def validate_distribution(base_dir, class_names=["good", "bad"]):
    """
    Check number of files for each class in each client folder.

    Args:
        base_dir (str): Base directory containing client folders (client_1, client_2, etc.)
        class_names (list): List of subfolder names to count files in (e.g., ["good", "bad"])
    """
    print(f"Validating dataset distribution in: {base_dir}")
    clients = sorted([f for f in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, f))])

    for client in clients:
        client_path = os.path.join(base_dir, client)
        print(f"\nğŸ“‚ {client}")
        total = 0
        for cls in class_names:
            cls_path = os.path.join(client_path, cls)
            if not os.path.exists(cls_path):
                print(f"   âš ï¸  {cls}: [MISSING]")
                continue
            count = len([f for f in os.listdir(cls_path) if f.endswith('.wav')])
            print(f"   âœ… {cls}: {count} files")
            total += count
        print(f"   ğŸ“Š Total files: {total}")
```

```python
base_dir = "resources/material/train-data/federated/balanced"  # adjust as per your structure
validate_distribution(base_dir)
```
