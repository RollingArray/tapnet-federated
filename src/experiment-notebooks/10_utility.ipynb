### âœ… Print Both Files and Folders

```python
import os

def print_directory_structure(base_path, indent=""):
    for entry in os.listdir(base_path):
        full_path = os.path.join(base_path, entry)
        if os.path.isdir(full_path):
            print(f"{indent}ğŸ“ {entry}/")
            print_directory_structure(full_path, indent + "    ")
        else:
            print(f"{indent}ğŸ“„ {entry}")

# Example usage
print_directory_structure("./")  # Replace './' with your target path
```

### âœ… Print Only Folders

```python
import os

def print_folders_only(base_path, indent=""):
    for entry in os.listdir(base_path):
        full_path = os.path.join(base_path, entry)
        if os.path.isdir(full_path):
            print(f"{indent}ğŸ“ {entry}/")
            print_folders_only(full_path, indent + "    ")

# Example usage
print_folders_only("../../src/experiment-notebooks")
```

### âœ… Print Only Files

```python
import os

def print_files_only(base_path, indent=""):
    for entry in os.listdir(base_path):
        full_path = os.path.join(base_path, entry)
        if os.path.isfile(full_path):
            print(f"{indent}ğŸ“„ {entry}")
        elif os.path.isdir(full_path):
            print_files_only(full_path, indent + "    ")

# Example usage
print_files_only("../../src/experiment-notebooks")
```
