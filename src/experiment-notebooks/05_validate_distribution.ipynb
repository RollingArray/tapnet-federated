# âœ… Validate Federated Data Distribution
This notebook verifies that the data distribution across clients (IID or Non-IID) is as expected using label counts, pie charts, and statistical summaries.

---

```python
import os

def validate_distribution(base_dir, mode="IID", file_type="wav", class_names=["good", "bad"]):
    """
    Validate dataset distribution per client.

    Args:
        base_dir (str): Root directory like "../../resources/material/train-data/federated"
        mode (str): "IID", "non-IID", "IID-npy", "non-IID-npy"
        file_type (str): "wav" or "npy"
        class_names (list): Class subfolders to validate for each client
    """
    full_path = os.path.join(base_dir, mode)
    file_ext = f".{file_type.lower()}"

    print(f"Validating dataset distribution in: {full_path} ({file_type.upper()}, {mode})")
    
    if not os.path.exists(full_path):
        print(f"âŒ Directory not found: {full_path}")
        return

    clients = sorted([f for f in os.listdir(full_path) if os.path.isdir(os.path.join(full_path, f))])

    for client in clients:
        client_path = os.path.join(full_path, client)
        print(f"\nğŸ“‚ {client}")
        total = 0

        if client == "test_clients":
            # Detect if it's flat or nested
            contents = os.listdir(client_path)
            subdirs = [d for d in contents if os.path.isdir(os.path.join(client_path, d))]
            npy_count = 0

            if set(class_names).intersection(subdirs):  # Nested: good/bad
                for cls in class_names:
                    cls_path = os.path.join(client_path, cls)
                    if os.path.exists(cls_path):
                        count = len([f for f in os.listdir(cls_path) if f.endswith(file_ext)])
                        print(f"   âœ… {cls}: {count} files")
                        npy_count += count
                    else:
                        print(f"   âš ï¸  {cls}: [MISSING]")
                print(f"   ğŸ“¦ {file_ext} files: {npy_count} files (nested folder)")
            else:
                # Flat folder
                npy_count = len([f for f in contents if f.endswith(file_ext)])
                print(f"   ğŸ§ {file_ext} files: {npy_count} files (flat folder)")

            print(f"   ğŸ“Š Total files: {npy_count}")
            continue

        # For regular clients with good/bad structure
        for cls in class_names:
            cls_path = os.path.join(client_path, cls)
            if not os.path.exists(cls_path):
                print(f"   âš ï¸  {cls}: [MISSING]")
                continue
            count = len([f for f in os.listdir(cls_path) if f.endswith(file_ext)])
            print(f"   âœ… {cls}: {count} files")
            total += count

        print(f"   ğŸ“Š Total files: {total}")
```

### Validating IID wav dataset class distribution

```python
# Validate WAV data in IID mode
validate_distribution(
    base_dir="../../resources/material/train-data/federated",
    mode="IID",
    file_type="wav"
)
```

### Validating Non-IID wav dataset class distribution

```python
# Validate WAV data in IID mode
validate_distribution(
    base_dir="../../resources/material/train-data/federated",
    mode="non-IID",
    file_type="wav"
)
```

### Validating Non-IID npy dataset class distribution

```python
# Validate WAV data in IID mode
validate_distribution(
    base_dir="../../resources/material/train-data/federated",
    mode="non-IID-npy",
    file_type="npy"
)
```
