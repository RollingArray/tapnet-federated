# üéôÔ∏è Extract Tap Segments from Raw Audio
This notebook segments raw continuous audio files into individual tap events. It applies silence detection or energy thresholding to isolate tap sounds from background noise.

---

```python
import sys
import os
import librosa

# Add the 'src' directory to the Python path
sys.path.append(os.path.abspath('../../src'))

from core.tap_detection import detect_taps, plot_waveform_with_taps, extract_and_save_taps
```

### Extract individual good audio taps from a source audio

```python
# Configuration
MATERIAL_TYPE='good'
THRESHOLD = 0.04

# Define paths
audio_path = f'../../resources/material/train-data/{MATERIAL_TYPE}-material.wav'
output_dir = f'../../resources/material/train-data/{MATERIAL_TYPE}-material-taps'

# Load audio
y, sr = librosa.load(audio_path, sr=None)

# Detect taps
tap_times, tap_frames, rms = detect_taps(y, sr, threshold=THRESHOLD)

# Plot interactively
plot_waveform_with_taps(y, sr, tap_times, rms, THRESHOLD)

# Save if satisfied
SAVE = True
if SAVE:
    extract_and_save_taps(y, sr, tap_times, output_dir, material_type=MATERIAL_TYPE)
```
